```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import joblib  # For saving the model
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# ðŸ”¹ Load dataset
csv_file_path = "/mnt/data/SolarCellDB.csv"
df = pd.read_csv("/content/SolarCellDB.csv", low_memory=False)  # Fix dtype warning

# ðŸ”¹ Standardize column names (remove spaces & fix inconsistencies)
df.columns = df.columns.str.strip().str.lower()  # Convert to lowercase for consistency

# ðŸ”¹ Check available column names
print("Updated column names:", df.columns.tolist())  # Debugging step

# ðŸ”¹ Find the correct 'PCE' column dynamically
target_variable = None
for col in df.columns:
    if 'pce' in col.lower():  # Find any variation of 'PCE'
        target_variable = col
        break

if target_variable is None:
    raise KeyError("Could not find the 'PCE' column. Please check the dataset.")

print(f"Using '{target_variable}' as the target column.")  # Confirm the column name

# ðŸ”¹ Remove columns with all NaN, empty values, or False
df = df.dropna(axis=1, how='all')  # Drop columns where all values are NaN
df = df.loc[:, ~(df.isin([False]).all())]  # Remove columns where all values are False
df = df.loc[:, ~(df.eq('').all())]  # Remove columns where all values are empty

# ðŸ”¹ Convert mixed-type columns to strings to avoid dtype issues
for col in df.select_dtypes(include=['object']).columns:
    df[col] = df[col].astype(str)

# ðŸ”¹ Encode categorical features
categorical_cols = df.select_dtypes(include=['object']).columns
for col in categorical_cols:
    df[col] = LabelEncoder().fit_transform(df[col])

# ðŸ”¹ Define features and target variable
X = df.drop(columns=[target_variable])  # Features
y = df[target_variable]  # Target

# ðŸ”¹ Check for NaN values in y and remove them
if y.isna().sum() > 0:
    print(f"Warning: {y.isna().sum()} NaN values found in target variable. Removing them.")
    df = df.dropna(subset=[target_variable])  # Drop rows where target is NaN
    X = df.drop(columns=[target_variable])  # Update X after dropping NaN rows
    y = df[target_variable]  # Update y after dropping NaN rows

# ðŸ”¹ Alternative: Fill NaN values instead of dropping (Optional)
# y.fillna(y.mean(), inplace=True)  # Uncomment to replace NaN with mean instead of removing

# ðŸ”¹ Split dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# ðŸ”¹ Train Random Forest model
model = RandomForestRegressor(n_estimators=200, random_state=42, max_depth=15, n_jobs=-1)
model.fit(X_train, y_train)

# ðŸ”¹ Predict on test data
y_pred = model.predict(X_test)

# ðŸ”¹ Evaluate the model
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print(f"MAE: {mae:.4f}")
print(f"MSE: {mse:.4f}")
print(f"RMSE: {rmse:.4f}")
print(f"RÂ² Score: {r2:.4f}")

# ðŸ”¹ Save the model
joblib.dump(model, "trained_model.pkl")
print("Model saved as 'trained_model.pkl'")

# ðŸ”¹ Feature Importance Plot
feature_importances = model.feature_importances_
feature_names = X.columns
sorted_idx = np.argsort(feature_importances)[::-1]

plt.figure(figsize=(10, 6))
sns.barplot(x=feature_importances[sorted_idx][:10], y=np.array(feature_names)[sorted_idx][:10])
plt.xlabel("Feature Importance")
plt.ylabel("Feature")
plt.title("Top 10 Important Features")
plt.show()
```
